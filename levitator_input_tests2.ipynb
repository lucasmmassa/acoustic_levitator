```python
import numpy as np
import matplotlib.pyplot as plt
from time import sleep
from scipy import signal
```

## Amplitude A = 3

```python
inp = np.load('input_tests2/amplitude3/input.npy')
x = np.load('input_tests2/amplitude3/x.npy')
x = x-x[0]
y = np.load('input_tests2/amplitude3/y.npy')
y = y-y[0]

fig, ax = plt.subplots(1, 3)
fig.set_figheight(5)
fig.set_figwidth(15)

ax[0].plot(inp, 'orange')
ax[0].plot(inp, 'bo')

ax[1].plot(x, 'orange')
ax[1].plot(x, 'bo')

ax[2].plot(y, 'orange')
ax[2].plot(y, 'bo')
```
